<h2> Multiple Linear Regression </h2>  <br>
In multiple regression we have more than one independent variable and one dependent variable.
The solution here will be represented by a plane and the generic form is given as below.
    
   <center><i> Y = β<sub>0</sub> + β<sub>1</sub>X<sub>1</sub>+ β<sub>2</sub>X<sub>2</sub>+..... + β<sub>k</sub>X<sub>k</sub> + ε </i> </center> <br>
Multiple linear regression attempts to model the relationship between two or more feartures and a response by fitting a linear equation to observed data.The steps to perform multiple linaer regression are almost similar to that of simple linear regression.
The difference lies in the evaluation.You can use it to find out which factor has the highest impact on the predicted output and how different variables relate to each other.

```python
## Import the libraries
import pandas as pd
import numpy as np

from sklearn import linear_model
from sklearn import datasets
from sklearn.model_selection import train_test_split
```

```python
## Import the dataset
data = datasets.load_boston()
df = pd.DataFrame(data.data, columns = data.feature_names)
df.head()
```

```python
target = pd.DataFrame(data.target,columns=['MEDV'])
target.head()
```

<font face ='verdana'><font color = 'red'>Notice here, in following cell we are using more than one dependent variable in X</font>

```python
X= pd.DataFrame(df[['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX','PTRATIO', 'B', 'LSTAT']])
y = pd.DataFrame(target['MEDV'])
```

```python
X.head()
```

```python
#Split the data...
#split the data in 80/20 proportion
X_train ,X_test, y_train, y_test = train_test_split(X,y,test_size =0.2, random_state = 0)
```

```python
print('○ X : ',X.shape)
print("X_train :",X_train.shape)
print("X_test : ",X_test.shape)
print('\n○ y : ',y.shape)
print("y_train :",y_train.shape)
print("y_test: ",y_test.shape)
```

```python
lm = linear_model.LinearRegression()
model = lm.fit(X_train,y_train)
print('coeficent',lm.coef_)
print('Intercept',lm.intercept_)
print('R square',lm.score(X_train,y_train))
```

```python
y_pred = lm.predict(X_test)
print('y_pred : ',y_pred[0:5]) ## Printing only first five y_pred elements
print(len(y_pred))
```
